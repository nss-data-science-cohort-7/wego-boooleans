```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/processed_data.csv')
```

```python

# Set the visual settings for seaborn
sns.set_theme(style="whitegrid")

# Code for plotting the distribution of HDWY_DEV
plt.figure(figsize=(10, 6))
sns.histplot(df['HDWY_DEV'], bins=30, kde=True)
plt.title('Headway Deviation')
plt.xlabel('minutes')
plt.ylabel('trips')
plt.grid(False)
plt.show()

# Code for plotting the distribution of ADHERENCE means
# Group the data by UNIQUE_TRIP_ID to calculate the mean adherence per trip
grouped_data = df.groupby('UNIQUE_TRIP_ID')
adherence_mean = grouped_data['ADHERENCE'].mean()

# Plot the distribution of these means
plt.figure(figsize=(10, 6))
sns.histplot(adherence_mean, bins=30, kde=True)
plt.title('Adherence')
plt.xlabel('minutes')
plt.ylabel('trips')
plt.grid(False)
plt.show()
```
